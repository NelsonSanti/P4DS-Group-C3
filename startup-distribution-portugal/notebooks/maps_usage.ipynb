- plot_choropleth: Plots a Choropleth based on an aggregate of a provided column.

# Upgraded to Plot on District Maps

Pass in the `target="districts"`

````python
%run maps.ipynb #chart notebooks

import pandas as pd

data = {
    'distrito': ['Lisbon', 'Porto', 'Tomar', 'Coimbra', 'Braga', 'Évora', 'LEIRIA', 'LEIRIA', 'LEIRIA'],
    'income': [94.5, 88.2, 71.4, 79.8, 85.0, 64.1, 93.2, 45, 21.2]
}
df = pd.DataFrame(data)
plot_point_map(df, target="districts", label='distrito', metric='income', title="Income in Districts")
````

# Municipalities

Don't pass in a target to plot Municipalities, it is the default.
